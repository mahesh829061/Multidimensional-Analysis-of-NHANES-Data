```python
# Importing required libraries

import pandas as pd
import numpy as np
import seaborn as sns
```

```python
# Importing the data

diet = pd.read_csv('diet.csv')
exam = pd.read_csv('examination.csv')
labs = pd.read_csv('labs.csv')
```

```python
# Examining diet dataset

print("Diet Dataset")
print(diet.head())
print(" ")
print("Basic Info:")
print(diet.info())
print(" ")
print("Statistics")
print(diet.describe())
print(" ")
print("Missing Values:")
print(diet.isnull().sum())
```

```python
# Examining examination dataset

print("Examination Dataset")
print(exam.head())
print(" ")
print("Basic Info:")
print(exam.info())
print(" ")
print("Statistics")
print(exam.describe())
print(" ")
print("Missing Values:")
print(exam.isnull().sum())
```

```python
# Examining laboratory dataset

print("Lab Dataset")
print(labs.head())
print(" ")
print("Basic Info:")
print(labs.info())
print(" ")
print("Statistics")
print(labs.describe())
print(" ")
print("Missing Values:")
print(labs.isnull().sum())
```

```python
# Merging the three datasets using the common identifier (SEQN)

temp = pd.merge(diet, exam, on='SEQN', how='inner')
df = pd.merge(temp, labs, on='SEQN', how='inner')
df
```

```python
df.isnull().sum()
```

```python
df.fillna(df.mean(), inplace=True)
df
```

```python
df_with_null = df[df.columns[df.isnull().any()]].copy()
df_with_null
```

```python
df_with_null.isnull().sum()
```

```python
# Two columns have all their entries NaN so we shall remove those columns
df.drop(columns=['DRQSDT5','BMIHEAD'], inplace=True)
df_with_null.drop(columns=['DRQSDT5','BMIHEAD'], inplace=True)
```

```python
print(df_with_null.head())
print(df_with_null.isnull().sum())
```

```python
# Now we can see that only the columns with object dtypes have null values, we can't fill them with mean obviously so
# I will fill the null values with mode of the column

df.fillna(df.mode().iloc[0], inplace=True)
print(df.info())
print(df.isnull().sum())
```

Now, as for whether filling the null values with mean of the columns was a good idea or not. I would say that it was reasonable for columns with less percent of NaN because after all our data had information about medical characteristics of people, so taking the mean was not wrong.

However, many columns had more than half of their rows empty, so filling those columns too with the mean was a bad idea.

```python
# Now that we have filled the null values in our dataset
# We will now encode the columns with object datatypes
from sklearn.preprocessing import LabelEncoder

object_columns = df.select_dtypes(include=['object']).columns

label_encoder = LabelEncoder()
for column in object_columns:
    df[column] = label_encoder.fit_transform(df[column])

print(df.info())
```

```python
# Now we will scale our data using standard scaler

from sklearn.preprocessing import StandardScaler
numeric_columns = df.select_dtypes(include=['float64', 'int64', 'int32']).columns
scaler = StandardScaler()
df[numeric_columns] = scaler.fit_transform(df[numeric_columns])
df
```

Standard Scaling is important because of the following reasons:

1) Algorithms that rely on distances between data points, such as k-nearest neighbors (KNN) and hierarchical clustering, are sensitive to the scale of features. Standard scaling ensures that features contribute equally to the distance computations.
2) If features have vastly different scales, the algorithm that we use in the future may take longer to converge or may even fail to converge.
3) Without scaling, the magnitude of coefficients can be influenced by the scale of the features, making it difficult to compare their relative importance.
4) When features are on similar scales, it's easier to create meaningful visualizations that accurately represent the relationships within the data.
